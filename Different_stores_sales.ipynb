```python
df['invoice_time']=df['invoice_date'].apply(lambda x:x.split(' ')[1])
df['invoice_time']
#lambda x:: This is the syntax for creating an anonymous function (lambda function) in Python. The x is the input parameter to the function.
#x.split(' '): The input x is assumed to be a string, and this part of the lambda function uses the split method to split thestring into a list of substrings based on the space (' ') character.
#[1]: This part selects the second element of the list obtained from the split operation. In Python, indexing starts from 0, so [1] corresponds to the second element of the list. In the example above, it would select "9:58", which is the time part.
```

```python
df['invoice_date']=df['invoice_date'].apply(lambda x:x.split(' ')[0]) 
df['invoice_date']

#lambda: It is a keyword in Python used to create anonymous functions. 
#x:: This specifies the input parameter of the lambda function. In this case, x is the input variable.
#x.split(' ')[0]: This is the expression that the lambda function will evaluate. It takes the input x, which is assumed to be a
#string, and applies the split(' ') method. This method splits the string into a list of substrings using a space (' ') as the
#delimiter. The [0] at the end extracts the first element of the resulting list, which corresponds to the part before the first
#space.


```

```python
df['age_category']=pd.cut(df.age,[0,25,40,55,71], labels=["youth","adults","older","seniors"])


#df.age: This refers to the "age" column in your DataFrame (df), which presumably contains numerical values representing ages.
#pd.cut(): This function is used to segment and sort data values into bins. It takes three main arguments:
#The first argument is the data to be binned (df.age).
#The second argument is the bin edges, specifying the ranges for the bins. In this case, [0, 25, 40, 55, 71] defines four bins: 0-25, 25-40, 40-55, and 55-71.
#The third argument is the labels parameter, which assigns labels to the different bins. In your case, the labels are ["youth", "adults", "older", "seniors"].
#df["Age_category"]: This part of the code creates a new column in the DataFrame called "Age_category" and assigns the bin labels to each row based on the corresponding age value in the "age" column.
```

```python
# Grouping the DataFrame by 'invoice_date' and calculating the sum of 'total_cost' for each date.
daily_sales=df.groupby('invoice_date')['total_cost'].sum()

# Finding the date with the mazimum sales
max_sales_date=daily_sales.idxmax()

# Plotting the line plot
plt.figure(figsize=(12,6))
sns.lineplot(x=daily_sales.index, y=daily_sales.values, marker=0)
plt.title('Daily Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)

# Adding a vertical line at the date with the maximum sales
plt.axvline(x=max_sales_date,color='green',linestyle='--',label=f'Max Sales Date:{max_sales_date}')
plt.legend()
plt.show()

# Printing the date with the maximum sales
print(f"The date with maximum sales is: {max_sales_date}")
```
